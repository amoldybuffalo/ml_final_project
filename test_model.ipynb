```python
%matplotlib qt
import tensorflow as tf
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # sometimes needed to register 3D
import numpy as np
from matplotlib import colors
import h5py
from matplotlib.widgets import Slider

from medmnist import AdrenalMNIST3D

test_dataset = AdrenalMNIST3D(split='test', size=28, download=True)
testx = []
testy = []

for i in range(len(test_dataset)):
    testx.append(test_dataset[i][0])
    testy.append(test_dataset[i][1])

#validation
testx_tensor = tf.convert_to_tensor(testx, dtype=tf.float16)
testx_tensor = np.transpose(testx_tensor, (0,2,3,4,1))

testy_tensor = tf.convert_to_tensor(testy, dtype=tf.float16)

model = tf.keras.Sequential(
    layers = [
    tf.keras.layers.Input(shape=(28,28,28,1)),
    tf.keras.layers.Conv3D(16, 3, activation="relu"),
    tf.keras.layers.Conv3D(16, 2, activation="relu"),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
    ]
)
model.load_weights("model.ckpt.weights.h5")
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.evaluate(testx_tensor, testy_tensor)

```
